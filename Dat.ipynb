```python
import matplotlib 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#function to separate data
def separate_data(data):
    # Split the data into lines
    lines = data.split('\n')

    # Initialize empty lists for each column
    centerofmassenergy = []
    comerror = []
    sfactor = []
    sfactorerror = []

    # Iterate over each line, split by whitespace, and append to respective column lists
    for line in lines:
        values = line.split()
        centerofmassenergy.append(float(values[0]))
        comerror.append(float(values[1]))
        sfactor.append(float(values[2]))
        sfactorerror.append(float(values[3]))

    # Return the four columns as arrays
    return centerofmassenergy, comerror, sfactor, sfactorerror

```
